```python
# import directories
import pandas as pd
import numpy as np
```

```python
# import data
data = pd.read_csv("../data/sbdb_query_results.csv", low_memory=False)
```

```python
# separate the data into groups
trojans = data[data["class"] == "TJN"]
centaurs = data[data["class"] == "CEN"]
outermb = data[data["class"] == "OMB"]
```

```python
# if an asteroid is between the perihelion and aphelion it will likely be a trojan

perihelion = {"jupiter":5.4570}
aphelion = {"jupiter":4.9506}
```

```python
# eccentricity bounds
emax = 0.3
```

```python
import matplotlib.pyplot as plt
```

```python
# plots predicted trojans

plt.scatter(trojans["a"], trojans["e"], c = "r", s = 4, label = "trojans") # trojan points
plt.scatter(centaurs["a"], centaurs["e"], c = "b", s = 4, label = "centaurs") 
plt.scatter(outermb["a"], outermb["e"], c = "g", s = 4, label = "outer-middle-belt")
plt.axvline(perihelion["jupiter"], c = "orange", linestyle = ":", label = "jp perihelion/aphelion")
plt.axvline(aphelion["jupiter"], c = "orange", linestyle = ":")
plt.axhline(emax, c = "purple", linestyle = ":", label = "eccentricity max")
plt.legend(loc = 'upper right')
plt.semilogx()
plt.title("Distance vs eccentricity")
plt.xlabel("Distance (au)")
plt.ylabel("Eccentricity of orbit")
plt.show()
```

```python
# try some different plots

plt.scatter(trojans["diameter"], trojans["e"], c = "r", s = 4, label = "trojans") # trojan points
plt.scatter(centaurs["diameter"], centaurs["e"], c = "b", s = 4, label = "centaurs") 
plt.scatter(outermb["diameter"], outermb["e"], c = "g", s = 4, label = "outer-middle-belt")
plt.legend(loc = 'upper right')
plt.semilogx()
plt.title("Distance vs Diameter")
plt.xlabel("Distance (au)")
plt.ylabel("Diameter")
plt.show()
```

```python
# try more parameters

plt.scatter(trojans["a"], trojans["i"], c = "r", s = 4, label = "trojans") # trojan points
plt.scatter(centaurs["a"], centaurs["i"], c = "b", s = 4, label = "centaurs") 
plt.scatter(outermb["a"], outermb["i"], c = "g", s = 4, label = "outer-middle-belt")
plt.legend(loc = 'upper right')
plt.semilogx()
plt.title("Distance vs Inclination")
plt.xlabel("Distance (au)")
plt.ylabel("Inclination (deg)")
plt.show()
```

### Analysis
This parameter tests the inclination, which is the angle with respect to the x-y eclpitic plane, with respect to distance. The overall trend seems to be the same as everything else. Trojans and the outer-middle-belt are more similar with the centaurs having more variation.

```python
# calculates predicted trojans using the aphelion and perihelion values of jupiter
predtrojans = trojans[(trojans["a"] >= aphelion["jupiter"]) & (trojans["a"] <= perihelion["jupiter"])]
predtrojans = predtrojans[predtrojans["e"] <= emax]
```

```python
# calculates the accuracy of our predicted trojans
accuracy = len(predtrojans) / len(trojans)
print(accuracy)
```
